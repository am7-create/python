#Nominall /OHE Encoding

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
#create a sample df with a categorial variable
df = pd.DataFrame({
    'color': ['Red', 'Blue', 'Green', 'Green', 'Red', 'Blue']
})
```

```python
df
```

```python
#create an instance of hot encoder
encoder = OneHotEncoder()
```

```python
encoder.fit_transform(df[['color']])
```

```python
#fit the encoder to the data frmse & transfrom to the categorial variable :
encoded = encoder.fit_transform([['color']])
```

```python
import pandas as pd 
pd.DataFrame(encoded.toarray(), columns=encoder.get_feature_names_out())
```

```python
encoder.get_feature_names_out()
```

```python
# Assign the result to the variable name you want to use later
encoded_df = pd.DataFrame(encoded.toarray(), columns=encoder.get_feature_names_out())
```
