```python
#used for distribution analysis HISTOGRAM
import random
import matplotlib.pyplot as plt
data = [random.randint(1, 10) for _ in range(100)]
plt.hist(data, bins=10) #histogram
plt.title("momo")
plt.show()
```

```python
# subplots
categories = ['mon', 'tuesday', 'wed', 'thu', 'fri'] # 5 items
sales = [10, 20, 30, 40, 50]                       # Match with 5 items

y1 = [10, 20, 30, 40, 50]
y2 = [60, 70, 80, 90, 100]

plt.figure(figsize=(10, 5)) 

# first plot: BAR 
plt.subplot(1, 2, 1) 
plt.bar(categories, sales)
plt.title("Daily Sales")

# second plot: SCATTER 
plt.subplot(1, 2, 2) 
plt.scatter(y1, y2)
plt.title("Sales Relationship")

plt.show()
```
